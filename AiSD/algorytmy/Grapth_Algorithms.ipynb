# Depth First Search


# Traversals

```python
tree = {
    8: [1],
    1: [7,6,2],
    7: [3],
    6: [4],
    4: [5]
}

visited = set()

def post_order_traversal( root):
    if root not in tree:
        print(root)
        return
    for v in tree[root]:
        if v not in visited:
            visited.add(v)
            post_order_traversal(v)
    print(root)

post_order_traversal(8)


```

## DFS


- wersja rekurencyjna


```python
graph = {0: [1,7], 1:[0,7], 7:[0,1,2,4], 2:[7,4], 4:[7,2,3,5], 3:[4,5,6], 5:[4,5,8], 6:[3,8], 8:[6,5], 9:[]}

visited = [False for _ in range(len(graph.keys()))]
def dfs(current_node):
    if visited[current_node] == True:
        return
    visited[current_node] = True
    for neighbour in graph[current_node]:
        dfs(neighbour)
    print(current_node)

# graf nie jest spÃ³jny, stÄ…d wywoÅ‚ujemy dfs na wszystkich kluczach
for k in graph.keys():
    if visited[k] == False: 
        dfs(k)
        print('===========')
```

- wersja iteracyjna

```python
graph = {0: [1,7], 1:[0,7], 7:[0,1,2,4], 2:[7,4], 4:[7,2,3,5], 3:[4,5,6], 5:[4,5,8], 6:[3,8], 8:[6,5], 9:[]}

visited = [False for _ in range(len(graph.keys()))]

# w wersji iteracyjnej korzystamy ze stacka
def iter_dfs(start_at):
    stack = [start_at]
    while stack != []:
        current_node = stack.pop()
        
        if visited[current_node] == False:
            print(current_node)
            
            visited[current_node] = True
            
            for neighbour in graph[current_node]:
                
                if visited[neighbour] == False:
                    stack.append(neighbour)

iter_dfs(start_at=0)
```
